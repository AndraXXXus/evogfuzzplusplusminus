```python
import os, tempfile
from tests4py.framework.systemtest import tests4py_test_oracle, TestResult

from evogfuzz.oracle import OracleResult

def oracle(inp: str):
    tmp = tempfile.NamedTemporaryFile(delete=False)
    result = OracleResult.UNDEF
    inp = str(inp)
    try:
        tmp.write(inp.encode())
        tmp.seek(0)
        result = framework.systemtest.tests4py_test_oracle(work_dir=Path("tmp", "pysnooper_2").absolute(), path=Path(tmp.name), verbose=False, diversity=False)
    finally:
        tmp.close()
        os.unlink(tmp.name)
        if result is TestResult.PASSING:
            return OracleResult.NO_BUG
        elif result is TestResult.FAILING:
            return OracleResult.BUG
        
        return OracleResult.UNDEF
```

```python
oracle("-otest.log\n-O\n-cint=repr,bool=str")
#oracle("-otest1.log")
```

```python
from tests4py.projects.pysnooper import grammar, is_valid_grammar
from fuzzingbook.GrammarFuzzer import GrammarFuzzer, tree_to_string

is_valid_grammar(grammar)
fuzzer = GrammarFuzzer(grammar)

for _ in range(10):
    inp = fuzzer.fuzz_tree()
    print(tree_to_string(inp))
    print(inp, oracle(inp))
```

```python
from evogfuzz.evogfuzz_class import EvoGFuzz

epp = EvoGFuzz(
    grammar=grammar,
    inputs=["-d1\n-T\n", "-otest.log\n-cint=str\n"],
    oracle=oracle,
    iterations=10
)
found = set()
found = epp.fuzz()
```

```python
print(f"Found {len(found)} unique failing inputs")

for inp in list(found)[:20]:
    print("_________")
    print(str(inp))
    print(oracle(inp))
```

```python
from evogfuzz.evogfuzz_class import EvoGGen

egg = EvoGGen(
    grammar=grammar,
    inputs=["-cint=str\n"],
    oracle=oracle,
    iterations=10,
    transform_grammar=True
)

g, f = egg.optimize()
```

```python
for inp in list(f)[:20]:
    print("_________")
    print(str(inp))
    print(oracle(inp))
```

```python
from fuzzingbook.ProbabilisticGrammarFuzzer import ProbabilisticGrammarFuzzer

pfuzzer = ProbabilisticGrammarFuzzer(g)
for i in range(50):
    inp = pfuzzer.fuzz()
    print(inp)
    print(i, oracle(inp))
```
